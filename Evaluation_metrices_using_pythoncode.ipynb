```python
import numpy as np
import matplotlib.pyplot as plt

# Generate Synthetic Data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add Bias Term (Intercept)
X_b = np.c_[np.ones((100, 1)), X]

# Train Linear Regression Model Using Normal Equation
theta_best = np.linalg.inv(X_b.T @ X_b) @ (X_b.T @ y)

# Make Predictions
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_pred = X_new_b @ theta_best

# Predictions on Training Data
y_train_pred = X_b @ theta_best

# Calculate Evaluation Metrics Using Formulas

# Mean Absolute Error (MAE)
mae = np.mean(np.abs(y - y_train_pred))

# Mean Squared Error (MSE)
mse = np.mean((y - y_train_pred) ** 2)

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# R-squared (RÂ²)
ss_tot = np.sum((y - np.mean(y)) ** 2)
ss_res = np.sum((y - y_train_pred) ** 2)
r2 = 1 - (ss_res / ss_tot)

# Adjusted R-squared
n = X.shape[0]
p = X.shape[1]
adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))

# Print Evaluation Metrics
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (RÂ²): {r2:.4f}")
print(f"Adjusted R-squared: {adjusted_r2:.4f}")

# Plot Predictions vs True Values
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.7, color='b', label='True Values')
plt.plot(X_new, y_pred, color='r', linestyle='--', label='Predicted Values')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Predicted vs True Values')
plt.legend()
plt.show()
```
